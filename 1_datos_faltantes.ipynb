# Datos Faltantes

```python
# Importar librerías necesarias
import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.impute import SimpleImputer, KNNImputer
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar el conjunto de datos
diabetes = load_diabetes(as_frame=True)
data = diabetes.data
data['target'] = diabetes.target
```

```python
data.sample(10)
```

```python
# Simular valores faltantes
np.random.seed(42)
data.loc[data.sample(frac=0.1).index, 'bmi'] = np.nan
data.loc[data.sample(frac=0.1).index, 'bp'] = np.nan
```

# Ejercicios
Contesta las siguientes preguntas. Para cada pregunta, deberás escribir código que demostrará cómo llegaste al resultado:

### 1. ¿Cuántos valores faltantes hay en cada columna?**

```python
# Contar valores faltantes por columna
missing_values = data.isnull().sum()
print(missing_values)
```

### 2. Utiliza imputación simple (media) para llenar los valores faltantes de la columna 'bmi'.

```python
# Crear una copia para la imputación simple
data_simple = data.copy()

# Configurar el imputador
imputer_simple = SimpleImputer(strategy='mean')

# Aplicar la imputación a la columna 'bmi'
# (Usamos [[ ]] para que scikit-learn reciba un DataFrame de 2D)
data_simple['bmi'] = imputer_simple.fit_transform(data[['bmi']])

print(f"Valores faltantes en bmi tras imputación simple: {data_simple['bmi'].isnull().sum()}")
```

### 3. Utiliza KNNImputer para imputar valores en las columnas 'bmi' y 'bp'. Compara los resultados con los de la imputación simple.

```python
# Crear una copia para KNN
data_knn = data.copy()

# Configurar KNNImputer (por defecto usa 5 vecinos)
imputer_knn = KNNImputer(n_neighbors=5)

# Aplicar a las columnas bmi y bp
data_knn[['bmi', 'bp']] = imputer_knn.fit_transform(data[['bmi', 'bp']])

# Comparación rápida de los primeros registros que eran NaN
print("Comparación de resultados (primeras filas):")
print(data_knn[['bmi', 'bp']].head())
```

### 4. Genera un histograma comparando los datos antes y después de la imputación en la columna 'bmi'.

```python
plt.figure(figsize=(10, 6))

# Histograma de datos originales (quitando los NaN para que no de error)
sns.histplot(data['bmi'].dropna(), kde=True, color="blue", label="Original (con huecos)", alpha=0.5)

# Histograma tras imputación simple
sns.histplot(data_simple['bmi'], kde=True, color="red", label="Imputación Simple (Media)", alpha=0.3)

# Histograma tras KNN
sns.histplot(data_knn['bmi'], kde=True, color="green", label="Imputación KNN", alpha=0.3)

plt.title('Comparación de Distribución: BMI')
plt.legend()
plt.show()
```
